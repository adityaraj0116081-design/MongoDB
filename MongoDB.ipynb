8. What is sharding in MongoDB? How does it differ from replication?


> Sharding in MongoDB is a technique used to horizontally partition (split) data across multiple servers called shards. It helps MongoDB handle very large datasets and high read/write traffic by distributing the workload.

How Sharding Works
. Data is divided based on a shard key.
. Each shard stores only a portion of the total data.
. A Config Server stores metadata about where data is located.
. A mongos router receives client requests and forwards them to the correct shard.

Replication in MongoDB

Replication is the process of maintaining multiple copies of the same data on different servers using a Replica Set.

A replica set consists of:

. Primary: Accepts all write operations.
. Secondary: Copies data from the primary.
. If the primary fails, a secondary is automatically elected as the new primary.

Replication provides:

. High availability
. Fault tolerance
. Data backup

10.  What are the ACID properties in the context of MongoDB transactions?

>ACID properties ensure that MongoDB transactions are reliable, consistent, and safe, especially when multiple operations need to succeed or fail together.

ACID Properties in MongoDB

1. Atomicity (A)

A transaction is treated as a single unit.
Either all operations succeed or none of them are applied.
If one operation fails, the entire transaction is rolled back.

Example:
If transferring ₹1000 from Account A to Account B:

Deduct ₹1000 from A.
Add ₹1000 to B.
If adding to B fails, the deduction from A is also undone.

2. Consistency (C)

A transaction brings the database from one valid state to another.
Database rules, constraints, and data integrity are always maintained.

Example:
The total money in the banking system remains the same before and after a successful transfer.

3. Isolation (I)

Transactions execute independently.
One transaction cannot see the incomplete changes made by another transaction.

Example:
If one user is updating a document, another user will not see the partially updated data until the transaction is committed.

4. Durability (D)

Once a transaction is committed, its changes are permanently saved.
The data remains safe even if the server crashes immediately after the commit.

Example:
After a successful transaction, the updated data is not lost due to a power failure or system restart.

Summary Table
Property	Meaning
Atomicity	All operations succeed or all fail.
Consistency	Database remains in a valid state before and after the transaction.
Isolation	Transactions do not interfere with each other.
Durability	Committed data is permanently stored.

19. How do MongoDB transactions differ from SQL transactions?

> MongoDB and SQL databases both support ACID transactions, but they differ in their design, data model, and typical usage.

Difference Between MongoDB Transactions and SQL Transactions

| Feature                   | MongoDB Transactions                      | SQL Transactions                                                            |
| ------------------------- | ----------------------------------------- | --------------------------------------------------------------------------- |
| Database Type             | NoSQL (Document-oriented)                 | Relational (Table-based)                                                    |
| Data Model                | BSON documents                            | Tables with rows and columns                                                |
| Default Atomicity         | Single-document operations are atomic     | Individual SQL statements are atomic                                        |
| Multi-Record Transactions | Supported (MongoDB 4.0+)                  | Supported by default                                                        |
| Joins                     | Limited (`$lookup` in aggregation)        | Rich support using `JOIN`                                                   |
| Schema                    | Flexible (schema-less)                    | Fixed schema                                                                |
| Performance               | Best when transactions are used sparingly | Optimized for frequent complex transactions                                 |
| Scalability               | Horizontal scaling using sharding         | Primarily vertical scaling (though some systems support horizontal scaling) |


20. What are the main differences between capped collections and regular collections?

> Capped collections are fixed-size collections that automatically overwrite the oldest documents when they become full, whereas regular collections can grow dynamically as more data is added.

Difference Between Capped Collections and Regular Collections

| Feature          | Capped Collection                              | Regular Collection                        |
| ---------------- | ---------------------------------------------- | ----------------------------------------- |
| Size             | Fixed size                                     | Grows dynamically                         |
| Document Removal | Oldest documents are automatically overwritten | Documents remain until explicitly deleted |
| Insertion Order  | Preserves insertion order                      | No guaranteed insertion order             |
| Storage          | Pre-allocated                                  | Allocated as needed                       |
| Performance      | Fast inserts and reads                         | General-purpose performance               |
| Deletion         | Automatic when full                            | Manual deletion required                  |

```python
# 1.Write a Python script to load the Superstore dataset from a CSV file into MongoDB.

import pandas as pd
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Create or connect to database
db = client["superstore_db"]

# Create or connect to collection
collection = db["orders"]

# Load CSV file
csv_file = "Superstore.csv"   # Replace with your CSV file path
df = pd.read_csv(csv_file, encoding="latin1")

# Convert date columns (if present)
date_columns = ["Order Date", "Ship Date"]

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Replace NaN values with None
df = df.where(pd.notnull(df), None)

# Convert DataFrame to dictionary
records = df.to_dict(orient="records")

# Insert data into MongoDB
if records:
    collection.insert_many(records)
    print(f"{len(records)} records inserted successfully!")
else:
    print("No records found in the CSV file.")

# Close MongoDB connection
client.close()
```

```python
# 9.Use aggregation to group orders by Region and calculate total sales per region.

import pandas as pd

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Select database and collection
db = client["Superstore"]
orders = db["Orders"]

# Aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": "$Region",
            "TotalSales": {"$sum": "$Sales"}
        }
    },
    {
        "$sort": {"TotalSales": -1}
    }
]

# Execute aggregation
results = orders.aggregate(pipeline)

# Display results
print("Total Sales by Region:")
for result in results:
    print(f"Region: {result['_id']}, Total Sales: {result['TotalSales']:.2f}")



```

```python
# 11. Count the number of orders for each category.

import pandas as pd
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Select database and collection
db = client["Superstore"]
orders = db["Orders"]

# Aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": "$Category",
            "OrderCount": {"$sum": 1}
        }
    },
    {
        "$sort": {"OrderCount": -1}
    }
]

# Execute aggregation
results = orders.aggregate(pipeline)

# Display results
print("Number of Orders by Category:")
for result in results:
    print(f"Category: {result['_id']}, Orders: {result['OrderCount']}")
```
